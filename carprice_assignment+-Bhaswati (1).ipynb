```python
################################ Business Rule ###########################################
# there is a business rule which states that there is a variable named CarName which is 
# comprised of two parts - the first word is the name of 'car company' and the second is the 
#'car model'. For example, chevrolet impala has 'chevrolet' as the car company name and 
#'impala' as the car model name

com= carprice["CarName"].str.split(" " , n =1, expand = True)
com1= carprice["CarName"].str.split(" ", n =2, expand = True)

carprice["CompanyName"]= com[0] 
carprice["CarName"]= com1[1]
#carprice[['CompanyName','CarName']]
carprice
```

```python
#########################  Data Treatment ##################################################
# If we look into unique values of car names, it seems some values are similar but with a different spelling 
# like maxda and mazda,nissan and Nissan,toyota and toyouta,vokswagen and volkswagen and vw, porsche and porcshce
# So, lets merge them

carprice[carprice['CompanyName'] =='maxda']
carprice['CompanyName']=carprice['CompanyName'].replace('maxda','mazda')
carprice

carprice[carprice['CompanyName'] =='Nissan']
carprice['CompanyName']=carprice['CompanyName'].replace('Nissan','nissan')
carprice

carprice[carprice['CompanyName'] =='toyouta']
carprice['CompanyName']=carprice['CompanyName'].replace('toyouta','toyota')
carprice

carprice[carprice['CompanyName'] =='vokswagen']
carprice['CompanyName']=carprice['CompanyName'].replace('vokswagen','volkswagen')
carprice

carprice[carprice['CompanyName'] =='vw']
carprice['CompanyName']=carprice['CompanyName'].replace('vw','volkswagen')
carprice

carprice[carprice['CompanyName'] =='porcshce']
carprice['CompanyName']=carprice['CompanyName'].replace('porcshce','porsche')
carprice
# If we look into unique values of drivewheel, it seems two values are similar but with different nomenclature 
# like fwd and 4wd. Lets merge both together and keep fwd

carprice[carprice['drivewheel'] =='4wd']
carprice['drivewheel']=carprice['drivewheel'].replace('4wd','fwd')
carprice
```

```python
#Step 3: Data Preparation
#You can see that your dataset has many columns with values as 'Yes' or 'No'.
#But in order to fit a regression line, we would need numerical values and not string. Hence, we need to convert them to 1s and 0s, where 1 is a 'Yes' and 0 is a 'No'.
# List of variables to map

varlist =  ['fueltype', 'aspiration', 'doornumber', 'enginelocation', 'drivewheel']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
carprice[varlist] = carprice[varlist].apply(binary_map)
```
